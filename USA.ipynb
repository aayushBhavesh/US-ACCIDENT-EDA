```python
sns.set_style("darkgrid")

plt.figure(figsize=(8,5))

sns.histplot(
    df['Start_Time'].dt.dayofweek,
    bins=7,
    stat="probability",
    discrete=True
)

plt.title("Distribution of Accidents by Day of Week")
plt.xlabel("Day of Week (0=Mon, 6=Sun)")
plt.ylabel("Proportion of Accidents")

plt.savefig("eda_plots/accidents_by_dayofweek_hist.png", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```

```python
sns.set_style("darkgrid")

# Aggregate
day_counts = (
    df['Start_Time']
    .dt.dayofweek
    .value_counts()
    .sort_index()
)

# Map to readable labels
day_map = {0:"Mon",1:"Tue",2:"Wed",3:"Thu",4:"Fri",5:"Sat",6:"Sun"}
day_counts.index = day_counts.index.map(day_map)
```

```python
plt.figure(figsize=(8,5))
day_counts.plot(kind='bar')

plt.title("Accidents by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Accidents")

plt.savefig("eda_plots/accidents_by_dayofweek.png", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```

```python
sns.set_style("darkgrid")

# Filter Sundays
sundays = df.loc[df['Start_Time'].dt.dayofweek == 6, 'Start_Time']

plt.figure(figsize=(8,5))

sns.histplot(
    sundays.dt.hour,
    bins=24,
    stat="probability",
    discrete=True
)

plt.title("Distribution of Accidents by Hour on Sundays")
plt.xlabel("Hour of Day")
plt.ylabel("Proportion of Accidents")

plt.savefig("eda_plots/sunday_accidents_by_hour.png", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```

```python
sns.set_style("darkgrid")

# Filter Mondays
monday_hours = (
    df.loc[df['Start_Time'].dt.dayofweek == 0, 'Start_Time']
    .dt.hour
    .value_counts()
    .sort_index()
)

plt.figure(figsize=(9,5))
monday_hours.plot(kind='bar')

plt.title("Monday Accidents by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")

plt.savefig("eda_plots/monday_accidents_by_hour.png", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```

```python
# ---------- FILTER ----------
df_2021 = df[df['Start_Time'].dt.year == 2021]

# Be robust with Source filtering (avoid exact-match traps)
df_2021_mapquest = df_2021[
    df_2021['Source'].str.contains("MapQuest", na=False)
]

# ---------- AGGREGATE ----------
monthly_counts = (
    df_2021_mapquest
    .dropna(subset=['Start_Time'])
    ['Start_Time']
    .dt.month
    .value_counts()
    .sort_index()
)
```

```python
# ---------- GUARD CLAUSE ----------
if monthly_counts.empty:
    print("âŒNo accident records are available for the MapQuest source in 2019. This indicates that MapQuest data collection either began after 2019 or is not covered for that year in the dataset. As a result, month-wise trend analysis for MapQuest in 2021 is not applicable.")
else:
    plt.figure(figsize=(9,5))
    monthly_counts.plot(kind='bar')

    plt.title("2021 Accidents by Month (MapQuest)")
    plt.xlabel("Month")
    plt.ylabel("Number of Accidents")

    plt.savefig(
        "eda_plots/2021_mapquest_accidents_by_month.png",
        dpi=300,
        bbox_inches="tight"
    )
    plt.show()
    plt.close()
```

```python
from folium.plugins import HeatMap
from datetime import datetime
from IPython.display import display
```

```python
# Ensure coordinates are valid
geo_df = (
    df[['Start_Lat', 'Start_Lng']]
    .dropna()
)

# Optional: remove invalid coordinates (important)
geo_df = geo_df[
    (geo_df['Start_Lat'].between(-90, 90)) &
    (geo_df['Start_Lng'].between(-180, 180))
]

# -------------------- SAMPLE --------------------
# Use FRACTION, not magic numbers
sample_df = geo_df.sample(frac=0.001, random_state=42)

lat_lon_pairs = sample_df.values.tolist()

# -------------------- CREATE MAP --------------------
m = folium.Map(
    location=[
        sample_df['Start_Lat'].mean(),
        sample_df['Start_Lng'].mean()
    ],
    zoom_start=5,
    tiles="CartoDB positron"
)

HeatMap(
    lat_lon_pairs,
    radius=8,
    blur=12,
    min_opacity=0.3
).add_to(m)

# -------------------- SAVE MAP --------------------
filename = f"eda_plots/accident_heatmap_{datetime.now().strftime('%Y%m%d_%H%M%S')}.html"
m.save(filename)

# -------------------- DISPLAY --------------------
display(m)
```
